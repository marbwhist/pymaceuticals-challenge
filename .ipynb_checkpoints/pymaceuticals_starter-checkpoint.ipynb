## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen



# study_data_complete_df.agg({'Tumor Volume (mm3)': ['min', 'max', 'median', 'skew']})

# study_data_complete_df[["Drug Regimen", "Tumor Volume (mm3)"]].groupby("Tumor Volume (mm3)").mean()


mean =study_data_complete_df.groupby("Drug Regimen")["Tumor Volume (mm3)"].mean()
median =study_data_complete_df.groupby("Drug Regimen")["Tumor Volume (mm3)"].median()
variance = study_data_complete_df.groupby("Drug Regimen")["Tumor Volume (mm3)"].var()
# standard_dev=study_data_complete_df.loc[:,"Score1"].std()
standard_dev=study_data_complete_df.groupby("Drug Regimen")["Tumor Volume (mm3)"].std()
sem=study_data_complete_df.groupby("Drug Regimen")["Tumor Volume (mm3)"].sem()

summary_data_complete_df=pd.DataFrame({"Mean": mean, "Median": median, "Variance": variance,
                                      "Standard Deviation": standard_dev,
                                      "SEM": sem})

summary_data_complete_df

# This method is the most straighforward, creating multiple series and putting them all together at the end.

```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# This method produces everything in a single groupby function

new_summary_stats=study_data_complete_df.groupby("Drug Regimen").agg({'Tumor Volume (mm3)':["mean", "median", "var", "std", "sem"]})


new_summary_stats
```
